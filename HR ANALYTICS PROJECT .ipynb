```python
# checking the gender gap

males1 = train.loc[train['gender'] == 'm', 'gender'].count()
females1 = train.loc[train['gender'] == 'f', 'gender'].count()
males2 = test.loc[test['gender'] == 'm', 'gender'].count()
females2 = test.loc[test['gender'] == 'f', 'gender'].count()

# Calculate the gender gap percentage for each file
total1 = males1 + females1
gap1 = abs((males1 - females1) / total1 * 100)
total2 = males2 + females2
gap2 = abs((males2 - females2) / total2 * 100)

# Plot the graph
labels = ['Train', 'Test']
values = [gap1, gap2]
plt.bar(labels, values)
plt.title('Gender Gap')
plt.xlabel('File')
plt.ylabel('Gap Percentage')
plt.show()
```

```python
# checking different regions of the company
print("different regions of the company\n",train['region'].unique())
```

```python
# scatter plot between average training score and is_promoted
train.columns
data=train
plt.scatter(data['avg_training_score'],data['is_promoted'])
plt.xlabel('Average Training Score',fontsize=20)
plt.ylabel('Is Promoted',fontsize=20)
plt.title('relationship between Average training Score and Promotion',fontsize=30)
plt.show()
```

```python
# checking dependency of different regions in promotion
data=train
table=pd.crosstab(data['region'],data['is_promoted'])
table.plot(kind='bar',stacked=True)
plt.xlabel('Region',fontsize=25)
plt.ylabel('Number of Employees',fontsize=25)
plt.title('Dependency of Regions in determining Promotion of Employees',color='black',fontsize=35)
plt.show()
```

```python
# Dependency of awards won on promotions
table2=pd.crosstab(train['awards_won?'],train['is_promoted'])
table2.plot(kind='bar',stacked=True)
plt.xlabel('awards_won?',fontsize=25)
plt.ylabel('Number of Employees',fontsize=25)
plt.title('Dependency of awards won on promotions',fontsize=35)
plt.show()
```

```python
# Dependency of Age factor in promotion of employees
table3=pd.crosstab(train['age'],train['is_promoted'])
table3.plot(kind='bar',stacked=True)
plt.xlabel('age',fontsize=25)
plt.ylabel('is_promoted',fontsize=25)
plt.title('Dependency of Age factor in Promotion of Employees',fontsize=35)
plt.show()
```

```python
# Checking which Department got the Greatest Number of Promotions
promotions=train.groupby(['department','is_promoted'])['is_promoted'].count().unstack()
promotions.plot(kind='bar')
plt.xlabel('department',fontsize=25)
plt.ylabel('is_promoted',fontsize=25)
plt.title(' Checking which Department got the Greatest Number of Promotions',fontsize=35)
plt.show()
```

```python
# Checking Dependency of Gender over promotion
table4=pd.crosstab(train['gender'],train['is_promoted'])
table4.plot(kind='bar',stacked=True)
plt.xlabel('gender',fontsize=25)
plt.ylabel('is_promoted',fontsize=25)
plt.title('Checking Dependency of Gender over promotion',fontsize=35)
plt.show()
```
